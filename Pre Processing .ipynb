#### Outliers 

```python
plt.boxplot(data['Fare'])
plt.title('Boxplot Of Fare Columns')
```

```python
#outlier present in the column Fare
```

```python
#handling outlier
Q1=np.percentile(data['Fare'],25,interpolation='midpoint')
Q2=np.percentile(data['Fare'],50,interpolation='midpoint')
Q3=np.percentile(data['Fare'],75,interpolation='midpoint')
```

```python
print(Q1)
print(Q2)
print(Q3)
```

```python
data['Fare'].median()
```

```python
IQR =Q3-Q1
```

```python
print(IQR)
```

```python
low_limit=Q1-1.5*IQR
up_limit=Q3+1.5*IQR
```

```python
print(low_limit)
print(up_limit)
```

```python
outlier=[]
for x in data['Fare']:
    if((x>up_limit)or(x<low_limit)):
        outlier.append(x)
```

```python
outlier
```

```python
index_out = data[(data['Fare']>up_limit)|(data['Fare']<low_limit)].index
```

```python
data.drop(index_out,inplace=True)
```
